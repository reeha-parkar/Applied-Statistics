### Seaborn (Working on the diabetes dataset)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
sns.lmplot(x='Age', y='BloodPressure', data=df)
plt.show()
```

```python
sns.lmplot(x='Age', y='Glucose', data=df, 
           fit_reg=False, # No regression line
           hue='Outcome' # color by evaluation stage
          )
plt.show()
```

```python
sns.swarmplot(x='Outcome', y='Glucose', data=df)
plt.show()
```

```python
sns.lmplot(x='Outcome', y='Glucose', data=df)
plt.show()
```

```python
sns.boxplot(data=df.loc[df['Outcome']==0, ['Glucose']])
plt.show()
```

```python
sns.boxplot(data=df.loc[df['Outcome']==1, ['Glucose']])
plt.show()
```

```python
df.corr
```

```python
sns.set(style='ticks', color_codes=True)
sns.pairplot(df)
plt.show()
```
