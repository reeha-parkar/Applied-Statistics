### Working with the tips dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data.head()
```

```python
tips_data.describe()
```

```python
sns.distplot(tips_data['total_bill'], kde=True).set_title('Histogram of total bill') # When kde=False,the curve of Gaussian is not visible 
plt.show()                                                                           # kde = Kernel Density Estimation
```

```python
sns.distplot(tips_data['total_bill'], kde=False) # Nota good practice 
sns.distplot(tips_data['tip'], kde=False).set_title('Histogram of total bill and tip')
plt.show()
```

```python
sns.boxplot(tips_data['total_bill']).set_title('Box plot of total bill')
plt.show()
```

```python
sns.boxplot(tips_data['tip']).set_title('Box plot of the tips received')
plt.show()
```

```python
sns.boxplot(tips_data['tip'])
sns.boxplot(tips_data['total_bill']).set_title('Box plot of total bill')
plt.show()
```

```python
sns.boxplot(x=tips_data['tip'], y=tips_data['smoker'])
plt.show()
```

```python
sns.boxplot(x=tips_data['tip'], y=tips_data['time'])
plt.show()

g = sns.FacetGrid(tips_data, row='time')
g = g.map(plt.hist, 'tip')
plt.show()
```

```python
g = sns.FacetGrid(tips_data, row='sex', col='smoker')
g = g.map(plt.hist, 'total_bill')
plt.show
```

```python
sns.boxplot(x=tips_data['tip'], y=tips_data['day'])
plt.show()

g = sns.FacetGrid(tips_data, row='day')
g = g.map(plt.hist, 'tip')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips_data)
plt.show()
```

```python
# Categoorical data
sns.catplot(x='sex', y='tip', data=tips_data)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips_data)
plt.show()
```

```python
sns.catplot(x='smoker', y='tip', order=['No', 'Yes'], data=tips_data)    # order also changes colour
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', data=tips_data)
plt.show()
```
